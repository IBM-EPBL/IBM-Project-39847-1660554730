```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = cv2.imread(r"D:\Plant Disease\Dataset Plant Disease\fruit-dataset\fruit-dataset\train\Apple___Black_rot\35f56736-d311-4e80-88a4-7ba12fd71985___JR_FrgE.S 2787.JPG")
b = cv2.imread(r"D:\Plant Disease\Dataset Plant Disease\fruit-dataset\fruit-dataset\train\Peach___Bacterial_spot\10db8907-e305-4623-943a-198e560ced42___Rut._Bact.S 0810.JPG")
```

```python
# show images
plt.subplot(121), plt.imshow(a)
plt.grid(False), plt.xticks([]), plt.yticks([])

plt.subplot(122), plt.imshow(b)
plt.grid(False), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Calculate histogram of both images for last channel.
hist_a = cv2.calcHist([a],[2],None,[256],[0,256])
hist_b = cv2.calcHist([b],[2],None,[256],[0,256])

# Plot histograms
plt.subplot(121)
plt.plot(hist_a)

plt.subplot(122)
plt.plot(hist_b)

plt.show()
```

```python
# plot cumulative histograms

cdf_b = hist_b.cumsum()
cdf_a = hist_a.cumsum()

plt.subplot(221), plt.plot(cdf_a), plt. title('cdf of a image')
plt.subplot(222), plt.plot(hist_a, 'k'), plt. title('pdf of a image')

plt.subplot(223), plt.plot(cdf_b), plt. title('cdf of b image')
plt.subplot(224), plt.plot(hist_b, 'k'), plt. title('pdf of b image')

# adjust the placement of subplots
plt.subplots_adjust(bottom=2, right=0.8, top=3)

plt.show()
```

```python
b_gray = cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)
a_gray = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
```

```python
# show images and their histograms
plt.subplot(221), plt.imshow(a_gray, cmap='gray')
plt.grid(False), plt.xticks([]), plt.yticks([])
plt.subplot(223), plt.plot(cv2.calcHist([a_gray],[0],None,[256],[0,256]))

plt.subplot(222), plt.imshow(b_gray, cmap='gray')
plt.grid(False), plt.xticks([]), plt.yticks([])
plt.subplot(224), plt.plot(cv2.calcHist([b_gray],[0],None,[256],[0,256]))

plt.show()
```
